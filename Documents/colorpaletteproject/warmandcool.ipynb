```python
import pandas as pd
import colorsys
```

```python
df = pd.read_csv('skincolors.csv')
```

```python
df.dropna(subset = ["W or C", "RGB", "category"])
```

```python
def findmax():
    # find max and min of warm and cool overall
    warmLOL = []
    coolLOL = []
    for index, row in df.iterrows():
        
        # if spreadsheet[row]["W or C"] == "warm":
        colors = row["RGB"]
        if type(colors) == str:
  
            colors = colors.replace(")", "")
            colors = colors.replace("(", "")
            r,g,b = colors.split(",")
            

            r = int(r)
            g = int(g)
            b = int(b)
        if row["W or C"] == "warm":
            warmLOL += [(colorsys.rgb_to_hsv(r, g, b))[0]]
        elif row["W or C"] == "cool":
            coolLOL += [(colorsys.rgb_to_hsv(r, g, b))[0]]
    print(max(warmLOL), "max warm")
    print(min(warmLOL), "min warm")
    print(max(coolLOL), "max cool")
    print(min(coolLOL), "min cool")
            # warmLOL += [spreadsheet[row]["RGB"]]
```

```python
findmax()
```

## finding what's warm and cool for dark or light

```python
def maxofdarklight(df):
    darkwarmhues = []
    darkcoolhues = []
    lightwarmhues = []
    lightcoolhues = []
    for index, row in df.iterrows():

            # if spreadsheet[row]["W or C"] == "warm":
            colors = row["RGB"]
            if type(colors) == str:
    
                colors = colors.replace(")", "")
                colors = colors.replace("(", "")
                r,g,b = colors.split(",")
                

                r = int(r)
                g = int(g)
                b = int(b)   
            h,s,v = colorsys.rgb_to_hsv(r, g, b)

            halfwaybt80and255 = (80+255)/2
            xmin = 80
            xmax = 255


            if 80 < v and v < halfwaybt80and255: #you are dark if you have low value 
                # print("dark")
                if row["W or C"] == "warm":
                    darkwarmhues += [h]
                elif row["W or C"] == "cool":
                    darkcoolhues += [h]
            elif v > halfwaybt80and255:
                # print("light")
                if row["W or C"] == "warm":
                    lightwarmhues += [h]
                elif row["W or C"] == "cool":
                    lightcoolhues += [h]
    
    print(f'the min - max warm value of dark skin is {min(darkwarmhues)} - {max(darkwarmhues)}')
 
    print(f'the min - max cool value of dark skin is {min(darkcoolhues)} - {max(darkcoolhues)}')
 
    print(f'the min - max warm value of light skin is {min(lightwarmhues)} - {max(lightwarmhues)}')

    print(f'the min - max cool value of light skin is {min(lightcoolhues)} - {max(lightcoolhues)}')
```

```python
maxofdarklight(df)
```
